```python
#import datframe of recipes 
import pandas as pd

df_recipes = pd.read_csv("./data/recipes.csv")
```

```python
#remove columns not used

col_removed_recipes = ['AuthorId','AuthorName','CookTime','PrepTime','TotalTime','Images', 'Keywords','DatePublished','Description','FatContent','SaturatedFatContent','CholesterolContent','SodiumContent','CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent','RecipeServings', 'RecipeYield', 'RecipeInstructions']
df_recipes.drop(col_removed_recipes, axis =1, inplace = True)
```

```python
df_recipes.drop('RecipeIngredientQuantities', axis = 1, inplace = True)
```

```python
df_recipes.dropna(subset=['AggregatedRating'], inplace = True)
```

```python
used_categories = ['Frozen Desserts', 'Cheesecake', 'Chicken', 'Pork', 'Meat', 'Pie']
df_recipes = df_recipes[df_recipes['RecipeCategory'].isin(used_categories)]
df_recipes
```

```python
ingredients = df_recipes['RecipeIngredientParts']
```

```python
ingredients = ingredients.str.replace('c\(|\)', '', regex=True)
```

```python
ingredients = ingredients.str.replace('"','', regex=True)
ingredients = ingredients.str.replace(' ','', regex=True)
```

```python
ingredients = ingredients.str.split(',')
```

```python
ingredients[647]
```

```python
ingredients = ingredients.reset_index(drop=True)
ingredients
```

```python
#dropped lines with less than 5 ingredients
for i in range(len(ingredients)):
    if (len(ingredients[i]) < 5) :
        print(i)
        ingredients.drop(i, inplace=True)

ingredients
```

```python
ingredients = ingredients.reset_index(drop=True)
```

```python
#Create a dictionary of Ingredients
dic_ingredients = {}
i = 0
for row in ingredients:
    ing_auxs = row
    for ing_aux in ing_auxs:
        if ing_aux not in dic_ingredients:
            dic_ingredients[ing_aux] = i
            i += 1
```

```python
dic_ingredients
```

```python
df_dic_ing = pd.DataFrame.from_dict(dic_ingredients, orient='index', columns=['Ingredients'])
df_dic_ing.index.name = 'Key'
df_dic_ing.to_csv('./data/dictionary_ingredients.csv',index=True)
```

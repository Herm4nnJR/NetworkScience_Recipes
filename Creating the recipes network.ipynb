# Creating a Recipe Network
### Nodes are Ingredients, Edges are recipes

```python
import pandas as pd

df_recipes = pd.read_csv("./data/recipes.csv")


col_removed_recipes = ['AuthorId','AuthorName','CookTime','PrepTime','TotalTime','Images', 'Keywords','DatePublished','Description','FatContent','SaturatedFatContent','CholesterolContent','SodiumContent','CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent','RecipeServings', 'RecipeYield', 'RecipeInstructions']
df_recipes.drop(col_removed_recipes, axis =1, inplace = True)

df_recipes.drop('RecipeIngredientQuantities', axis = 1, inplace = True)
df_recipes.dropna(subset=['AggregatedRating'], inplace = True)

used_categories = ['Frozen Desserts', 'Cheesecake', 'Chicken', 'Pork', 'Meat', 'Pie']
df_recipes = df_recipes[df_recipes['RecipeCategory'].isin(used_categories)]

df_recipes = df_recipes[df_recipes['ReviewCount']> 25]
df_recipes = df_recipes.reset_index(drop=True)
```

```python
ingredients = df_recipes['RecipeIngredientParts']
ingredients = ingredients.str.replace('c\(|\)', '', regex=True)
ingredients = ingredients.str.replace('"','', regex=True)
ingredients = ingredients.str.replace(' ','', regex=True)
ingredients = ingredients.str.split(',')
```

```python
#dropped lines with less than 5 ingredients
for i in range(len(ingredients)):
    if (len(ingredients[i]) < 5) :
        ingredients.drop(i, inplace=True)
```

```python
df_recipes['Ingredients'] = ingredients
df_recipes.dropna(subset=['Ingredients'], inplace = True)
df_recipes = df_recipes.reset_index(drop=True)
df_recipes.drop('RecipeIngredientParts', axis = 1, inplace = True)
```

```python
ingredients = pd.read_csv('./data/dictionary_ingredients.csv')
dic_ingredients = ingredients['Key'].to_dict()
```

```python
dic_ingredients
```

```python
df_recipes['RecipeId'] = range(800)
df_recipes
```

## Redes de receitas se conectam caso compartilhem ingredientes

```python
df_recipes['Ingredients'][0][0]
```

```python
## Creating Edges
df_edges = pd.DataFrame(columns = ['node1','node2','value'])
df_edges
```

```python
from tqdm import tqdm
import warnings

# Configurar o filtro para ignorar os avisos de FutureWarning
with warnings.catch_warnings():
    warnings.filterwarnings("ignore", category=FutureWarning)

    for ing in tqdm(range(len(dic_ingredients))):
        actual = df_recipes[df_recipes['Ingredients'].apply(lambda arr: dic_ingredients[ing] in arr)]
        actual = actual['RecipeId'].to_numpy()
        for i in tqdm(range(len(actual))):
            for j in range(i+1,len(actual)):
                new_data = {'node1': actual[i], 'node2': actual[j], 'value': 1}
                df_edges = df_edges.append(new_data, ignore_index=True)
```

```python
df_agrupado = df_edges.groupby(['node1','node2']).sum().reset_index()
```

```python
df_agrupado
```

```python
dic_recipes = df_recipes.set_index('RecipeId')['Name'].to_dict()
dic_recipes
```

```python
df_agrupado['node1_r'] = df_agrupado['node1'].replace(dic_recipes)
df_agrupado['node2_r'] = df_agrupado['node2'].replace(dic_recipes)
```

```python
df_agrupado
```

# Criando Grafo/Rede

```python
import networkx as nx
G= nx.Graph()
```

```python
for _, row in df_agrupado.iterrows():
    node1 = row['node1_r']
    node2 = row['node2_r']
    count = row['value']
    G.add_edge(node1, node2, weight=count)
```

```python
nx.write_gml(G,'grafo_recipes.gml')
```
